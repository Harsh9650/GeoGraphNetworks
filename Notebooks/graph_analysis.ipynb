```python
import json
import networkx as nx
import matplotlib.pyplot as plt
from networkx.readwrite import json_graph
```

# Great Britain's Largest Road Network: TQ RoadLink

```python
with open("Networks/TQ_RoadLink.json", "r") as f:  # Load the JSON file
    TQ_RoadLink = json.load(f)

# Convert back to a NetworkX graph
tq_roadlink_network = json_graph.node_link_graph(TQ_RoadLink)

plt.figure(figsize= (20, 10)) # figsize= (50, 30) for better view.

nx.draw(tq_roadlink_network, nx.get_node_attributes(tq_roadlink_network, 'pos') , with_labels = False, node_size = 0.1)
plt.show()
```

# Some basic graph analysis on the prepared networks.

**(1) TQ RoadLink is directed or undirected Graph**

```python
print("Is TQ Roadlink a directed graph?", tq_roadlink_network.is_directed())

# TQ RoadLink is a undirected Graph.
```

**(2) Number of Node and Edges in the Graph**

```python
print(f"Number of nodes in the TQ RoadLink road network are: {nx.number_of_nodes(tq_roadlink_network)}")
print(f"Number of edges in the TQ RoadLink road network are: {nx.number_of_edges(tq_roadlink_network)}")
```

**(3) Degree of Nodes: Number of edges that are incident to a node.**

```python
degrees = list(tq_roadlink_network.degree())
print("Node degrees:", degrees[:20])

# Results show, node 0 is associated with two edges in the tq_roadlink_network.
```

**(4) Path Analysis**

```python
# Randomly selected source and target node.

shortest_path = nx.shortest_path(tq_roadlink_network, source=23400, target=199999, weight='weight')
path_length = nx.shortest_path_length(tq_roadlink_network, source=23400, target=199999, weight='weight')

print(f"Length of the shortest path: {path_length} meters" )
```

**(5) Degree Centrality: is provided for each node as a measure of importance or influence of a node within a graph based on its connectivity.**

```python
degree_centrality = nx.degree_centrality(tq_roadlink_network)
print("Degree centrality of node 0:", degree_centrality[0])
```
